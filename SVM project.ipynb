```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
pd.DataFrame(np.c_[data.data , data.target])
```

```python
data.target_names
```

```python
dataset = pd.DataFrame(np.c_[data.data , data.target] , columns=[list(data.feature_names) + ["target"]])
```

```python
dataset
```

```python
dataset.info()
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y , test_size = 0.2 , random_state=2020)
```

#  TRAIN SUPPORT VECTOR MODEL

```python
from sklearn.svm import SVC
```

```python
sc = SVC(kernel = "rbf")
sc.fit(X_train,y_train)
```

```python
sc.score(X_test,y_test)
```

#  FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train,y_train)
```

```python
X_train_sc = sc.transform(X_train)
```

```python
X_test_sc = sc.transform(X_test)
```

#  SVC WITH KERNEL POLYNOMIAL

```python
sc1 = SVC(kernel = "poly" , degree=2)
sc1.fit(X_train_sc , y_train)
```

```python
sc1.score(X_test_sc , y_test)
```

# SVC WITH KERNEL LINEAR

```python
linear_class = SVC(kernel="linear")
linear_class.fit(X_train_sc , y_train)
```

```python
linear_class.score(X_test_sc,y_test)
```
