```python
data.drop(columns = "customerID" , inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data1 = data.select_dtypes(include = "object")
```

```python
data1.columns
```

```python
lb = OneHotEncoder(sparse_output=False , drop="first")
ar = lb.fit_transform(data1)
col = lb.get_feature_names_out()
```

```python
data = pd.DataFrame(ar , columns =col)
```

```python
data
```

```python
X = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1]
y.info()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y ,random_state = 10, test_size = .25)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(multi_class='multinomial')
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
ann.add(Dense(units = 64 , activation= "relu", input_shape = (26,)))
ann.add(Dense(units = 32 , activation= "relu"))
ann.add(Dense(units = 16 , activation= "relu"))
ann.add(Dense(units = 8 , activation= "relu"))
ann.add(Dense(units = 4 , activation= "relu"))
ann.add(Dense(units = 2 , activation= "relu"))
ann.add(Dense(units = 1 , activation= "sigmoid"))
```

```python
ann.compile(optimizer="adam",loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
ann.fit(X_train,y_train , batch_size=90,epochs=50)
```

```python
y_pred = ann.predict(X_test)
pred_data = []
for i in y_pred:
    if i[0]>0.5:
        pred_data.append(1)
    else:
        pred_data.append(0)

# y_pred_new = (y_pred > 0.5).astype(int)
        
```
