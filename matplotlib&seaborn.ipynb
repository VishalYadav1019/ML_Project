```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
a=np.array([1,2,3,5,5,7,4])
b=np.array([1,2,3,4,5,6,7])
plt.plot(a,b)
```

```python
plt.figure(figsize=(10,3))
plt.plot(a,color="red",label="A")
plt.plot(b,color="blue",label="B")
plt.legend()
```

```python
y = np.array([35, 25, 25, 15])
ab = ["Apples", "Bananas", "Cherries", "Dates"]
me = [0, 0.3, 0.3,0.2 ]

plt.pie(y, labels = ab, explode = me)
plt.show() 
```

```python
sect_A=np.array([67,65,96,45,36,74])
sect_B=np.array([78,63,96,85,67,88])
no=([1,2,3,4,5,6])
mean_A=np.mean(sect_A)
print("mean_A =",mean_A)
mean_B=np.mean(sect_B)
print("mean_B =",mean_B)
```

```python
plt.figure(figsize=(10,3))
plt.bar(sect_A,no,color="red",label="sect A")
plt.bar(sect_B,no,color="blue",label="sect B")
plt.plot([63.833,63.833,63.833,63.833,63.833,63.833],no,c="black",label="Mean A")
plt.plot([79.5,79.5,79.5,79.5,79.5,79.5],no,c="skyblue",label="Mean B")
plt.legend()
plt.show()
```
